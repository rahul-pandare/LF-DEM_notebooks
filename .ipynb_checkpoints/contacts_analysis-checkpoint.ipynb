```python
# function to plot different contacts wrt to all phi
# plotting small-small/ large-large or small-large contacts for all phis

# znet is - total number of contacts/ number of particles forming those contatcs

topDir      = "/Volumes/Rahul_2TB/high_bidispersity/"
#topDir      = "/media/rahul/Rahul_2TB/high_bidispersity/"

contFile    = 'contacts.txt'

NP          = [1000]

run         = {500:8,1000:4,2000:2,4000:1}

phi         = [0.70,0.71,0.72,0.73,0.74,0.75,0.76]

ar          = [1.0, 1.4, 1.8, 2.0, 4.0]

off = 100

def contactPlots(ptype, figsave = False):
    plt.clf()
    cmap = matplotlib.colormaps['viridis_r'] #color scheme
    
    #col1 is for total number of fric contacts
    #col2 is for total number of particles with fric contact
    if ptype == 'tt':
        col1 = 0
        plotTitle = 'Total Contacts'
    elif ptype == 'ss':
        col1 = 2
        plotTitle = 'Small-Small Contacts'
    elif ptype == 'sl':
        col1 = 4
        plotTitle = 'Small-Large Contacts'
    elif ptype == 'll':
        col1 = 6
        plotTitle = 'Large-Large Contacts'
    else:
        sys.exit('Invalid plot type')
    
    col2 = col1 + 1
    
    for ii in range(len(NP)):
        for j in range(len(phi)):
            phir = '{:.3f}'.format(phi[j]) if len(str(phi[j]).split('.')[1])>2 else '{:.2f}'.format(phi[j])

            totcon    =[]
            totconstd =[]

            for k in range(len(ar)):
                dataname=topDir+'NP_'+str(NP[ii])+'/phi_'+phir+'/ar_'+str(ar[k])+'/Vr_0.5'
                if os.path.exists(dataname):
                    temp1=[]
                    tempstd1=[]

                    for l in range (run[NP[ii]]):
                        if os.path.exists(f'{dataname}/run_{l+1}/{contFile}'):
                            conFile = open(f'{dataname}/run_{l+1}/{contFile}', 'r')
                            contData = np.loadtxt(conFile) # reading only column 3 which has particle size
                            conFile.close()
                            
                            znet_totcont=[a / b for a, b in zip(contData[off:,col1], contData[off:,col2]) if b != 0 ]
                            temp1.append(np.mean(znet_totcont))
                            tempstd1.append(np.var(znet_totcont))

                    totcon.append(np.mean(temp1))
                    totconstd.append(np.sqrt(np.mean(tempstd1)))

            plt.errorbar(ar, totcon, yerr=totconstd, fmt='--o', label=r'$\phi$' + f' = {phir}',color=cmap((j+1)/len(phi)), capsize=2, linewidth=1.2, elinewidth=0.5)

    plt.grid(which='Both', alpha=0.2)

    plt.title(plotTitle, fontsize=18)
    plt.xlabel(r'$\delta$', fontsize=14,fontstyle='italic')
    plt.ylabel(r'$\langle Z_{net} \rangle$', fontsize=14,fontstyle='italic')

    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.legend(fontsize=14)
    #plt.legend(fontsize=10,loc='lower right', labelspacing=1, borderpad=1, ncol =2)

    plt.tight_layout()
    if figsave:
        figFormat=".pdf"
        fig_save_path = "/Users/rahul/City College Dropbox/Rahul Pandare/CUNY/Research/Bidisperse Project/analysis/figures/contacts/"
        plt.savefig(fig_save_path+plotTitle+'_NP_'+str(1000)+'_'+figFormat, bbox_inches="tight",dpi=500)
    return

    plt.show()
```

```python
# plotting different contacts in the system
# total contacts on small particles and on large particles

topDir      = "/Volumes/Rahul_2TB/high_bidispersity/"
#topDir      = "/media/rahul/Rahul_2TB/high_bidispersity/"

contFile    = 'contacts.txt'

NP          = [1000]

run         = {500:8,1000:4,2000:2,4000:1}

#phi         = [0.70,0.71,0.72,0.73,0.74,0.76]
phi         = [0.76]


ar          = [1.0, 1.4, 1.8, 2.0, 4.0]

off = 100

cmap = matplotlib.colormaps['viridis_r'] #color scheme

for ii in range(len(NP)):
    for j in range(len(phi)):
        phir = '{:.3f}'.format(phi[j]) if len(str(phi[j]).split('.')[1])>2 else '{:.2f}'.format(phi[j])
        
        contOnSmall     =[]
        contOnSmallstd  =[]

        contOnlarge     =[]
        contOnlargestd  =[]
        
        for k in range(len(ar)):
            dataname=topDir+'NP_'+str(NP[ii])+'/phi_'+phir+'/ar_'+str(ar[k])+'/Vr_0.5'
            if os.path.exists(dataname):
                temp1=[]
                tempstd1=[]
                
                temp2=[]
                tempstd2=[]
                
                for l in range (run[NP[ii]]):
                    conFile = open(f'{dataname}/run_{l+1}/{contFile}', 'r')
                    contData = np.loadtxt(conFile) # reading only column 3 which has particle size
                    conFile.close()
                    
                    znet_contOnSmall=[a / b for a, b in zip(contData[off:,8], contData[off:,9]) if b != 0 ]
                    temp1.append(np.mean(znet_contOnSmall))
                    tempstd1.append(np.var(znet_contOnSmall))
                    
                    znet_contOnLarge=[a / b for a, b in zip(contData[off:,10], contData[off:,11]) if b != 0 ]
                    temp2.append(np.mean(znet_contOnLarge))
                    tempstd2.append(np.var(znet_contOnLarge))
                    
                contOnSmall.append(np.mean(temp1))
                contOnSmallstd.append(np.sqrt(np.mean(tempstd1)))
                
                contOnlarge.append(np.mean(temp2))
                contOnlargestd.append(np.sqrt(np.mean(tempstd2)))
                
        plt.errorbar(ar, contOnSmall, yerr=contOnSmallstd, fmt='--o', label='Contact on small particles', color=cmap((1)/len(ar)), capsize=2, linewidth=1.2, elinewidth=0.5)
        plt.errorbar(ar, contOnlarge, yerr=contOnlargestd, fmt='--o', label='Contact on large particles', color=cmap((2)/len(ar)), capsize=2, linewidth=1.2, elinewidth=0.5)


        plt.grid(which='Both', alpha=0.2)

        plt.title(fr"$\phi = {phir}$", fontsize=18)
        plt.xlabel(r'$\delta$', fontsize=14,fontstyle='italic')
        plt.ylabel(r'$\langle Z_{net} \rangle$', fontsize=14,fontstyle='italic')

        plt.xticks(fontsize=12)
        plt.yticks(fontsize=12)
        plt.legend(fontsize=14)
        plt.legend(fontsize=10,loc='upper right', labelspacing=1,borderpad=1)

        plt.tight_layout()
        if False:
            figFormat=".png"
            fig_save_path = "/Users/rahul/City College Dropbox/Rahul Pandare/CUNY/Research/Bidisperse Project/analysis/figures/contacts"
            plt.savefig(fig_save_path+'/contacts2_NP'+str(1000)+'_phi_'+phir+figFormat, bbox_inches="tight",dpi=500)

        plt.show()
```

```python
# Checking for non fric and sliding contact

#topDir      = "/media/rahul/Rahul_2TB/high_bidispersity/"
topDir      = "/Volumes/Rahul_2TB/high_bidispersity/"

intFile     = "int_random_seed_params_stress100r_shear.dat"

NP          = [1000]

run         = {500:8,1000:4,2000:2,4000:1}

phi         = [0.70,0.71,0.72,0.73,0.74,0.76,0.77,0.78,0.79,0.792]

ar          = [1.0, 1.4, 1.8, 2.0, 4.0]
#ar          = [2.0]

s=0
for ii in range(len(NP)):
    for j in range(len(phi)):
        phir = '{:.3f}'.format(phi[j]) if len(str(phi[j]).split('.')[1])>2 else '{:.2f}'.format(phi[j])
        for k in range(len(ar)):
            dataname=topDir+'NP_'+str(NP[ii])+'/phi_'+phir+'/ar_'+str(ar[k])+'/Vr_0.5'
            if os.path.exists(dataname):
                print('/phi_'+phir+'/ar_'+str(ar[k]))
                for l in range (run[NP[ii]]):
                    hashCounter=0
                    temp=[]
                    contactList=[] # list with interaction parameters for each element at each timestep

                    interFile = open(f'{dataname}/run_{l+1}/{intFile}', 'r')
                    fileLines = interFile.readlines()[27:] # skipping the comment lines
                    for line in fileLines:
                        if not line.split()[0] == '#':
                            lineList = [float(value) for value in line.split()]
                            temp.append(lineList)
                        else:
                            hashCounter += 1 # checking if counter reaches 7 (7 lines of comments after every timestep data)
                            if hashCounter == 7: 
                                contactList.append(np.array(temp))
                                temp = []
                                hashCounter = 0
                    interFile.close()
                    
                    for sampleList in contactList:
                        for i in range (sampleList.shape[0]):
                            if int(sampleList[i,10]) == 1:
                                print(f'no fric cont - {dataname}/run_{l+1}')
                                s+=1
                            if int(sampleList[i,10]) == 3:
                                print(f'slid fric cont - {dataname}/run_{l+1}')
                                s+=1
if s ==0:
    print('\n All contatcs are non-sliding frictional')
```
