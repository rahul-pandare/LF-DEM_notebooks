```python
# 2nd attempt

#topDir = "/media/rahul/Rahul_2TB/high_bidispersity/NP_1000/phi_0.74/ar_2.0/Vr_0.5/run_2/"
topDir = "/Volumes/Rahul_2TB/high_bidispersity" #NP_1000/phi_0.74/ar_2.0/Vr_0.5/run_2/"

# user inputs:
dr     = .1
dtheta = 5

NP   = 1000
phi  = 0.74
ar   = 2.0
run  = 1
phii = '{:.3f}'.format(phi) if len(str(phi).split('.')[1]) > 2 else '{:.2f}'.format(phi)

Dir = topDir + "/NP_" + str(NP) + "/phi_"+ phii + "/ar_" + str(ar) + "/Vr_0.5/run_" + str(run) +"/"

datFile = glob.glob(Dir + 'data_*')[0]
parFile = glob.glob(Dir + 'par_*')[0]

NP = int(np.genfromtxt(datFile, skip_header=1, max_rows=1, comments = '_')[2])

lx = np.genfromtxt(datFile, skip_header=3, max_rows=1, comments = '_')[2]
lz = np.genfromtxt(datFile, skip_header=5, max_rows=1, comments = '_')[2]

t, gamma, gammapt, _,  _, _, _, _, _, _, \
_, _, _, minGap, _, _, _, _, _, _,       \
_, _, _, _,  _, _, _, _, _, _            \
= np.loadtxt(datFile, skiprows=37).transpose()

dtheta *= np.pi/180
rmin = 1 + np.min(minGap)

rbin = np.arange(rmin, np.max([lx,lz])/2. + dr,dr)
#rbin_width = rbin[1] - rbin[0]
rbin_edges = np.concatenate(([rbin[0] - dr / 2],rbin + dr / 2))

thetabin = np.arange(-np.pi, np.pi + dtheta,dtheta)
#dtheta = thetabin[1]-thetabin[0]

#deltar = rbin[2]-rbin[1]

histMat = np.zeros((len(rbin), len(thetabin)))

for i,mat in enumerate(parList[100:101]):
    radiusList = mat[:,1]
    xlist = mat[:,2]
    zlist = mat[:,3]
    xdist, zdist = xlist-xlist[:,np.newaxis], zlist-zlist[:,np.newaxis]
    
    # lees edwards boundary:
    xdist[zdist > lz/2.]  -= gamma[i]*lx
    zdist[zdist > lz/2.]  -= lz
    
    xdist[zdist < -lz/2.] += gamma[i]*lx
    zdist[zdist < -lz/2.] += lz
    
    # x peridodic:
    xdist[xdist >  lx/2.] -= lx
    xdist[xdist < -lx/2.] += lx
    
    dij = np.sqrt(xdist**2+zdist**2)
    tij = np.arctan2(zdist, xlist)
    
    for ij in range(len(rbin[0:-1])):
        condn = np.logical_and(dij>=rbin[ij],dij<(rbin[ij]+dr))
        #tij1 = np.full_like(tij, np.nan)
        tij1 = tij[condn]
        theta_surf = dtheta/2 * (2*rbin[ij]*dr + dr**2) #different from michel's code
        for ik in range(len(thetabin[0:-1])):
            condn2 = np.logical_and(tij1>=thetabin[ik], tij1< (thetabin[ik]+dtheta))
            histMat[ij,ik] += np.sum(condn2)/NP/theta_surf
            
#txtFile = open('Dir' + 'PDF_g_r_theta.txt', 'w')
#txtFile.write('# Theta bins')
                
histMat /= len(parList[100:101])
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Wedge

# Define the number of angular and radial bins
num_theta_bins = 10
num_r_bins = 5

# Generate example data
theta_bins = np.linspace(0, 2 * np.pi, num_theta_bins + 1)  # Angular bin edges
r_bins = np.linspace(0, 1, num_r_bins + 1)  # Radial bin edges

# Create a 2D array of values for each bin
values = np.random.rand(num_r_bins, num_theta_bins)  # Example values (random)

# Create the radial plot
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})

# Plot the data using wedges
for i in range(num_r_bins):
    for j in range(num_theta_bins):
        theta_start = np.degrees(theta_bins[j])
        theta_end = np.degrees(theta_bins[j + 1])
        r_start = r_bins[i]
        r_end = r_bins[i + 1]

        # Convert radial bins to degrees for width
        wedge = Wedge(
            (0, 0), r_end, theta_start, theta_end,
            width=r_end - r_start,
            facecolor=plt.cm.viridis(values[i, j]),
            edgecolor='k',
            transform=ax.transData._b
        )
        ax.add_patch(wedge)

# Set the radial limits
ax.set_ylim(0, r_bins[-1])

# Add a color bar
sm = plt.cm.ScalarMappable(cmap='viridis', norm=plt.Normalize(vmin=values.min(), vmax=values.max()))
sm.set_array([])
fig.colorbar(sm, ax=ax)

# Customize the plot
ax.set_xticks([])  # Turn off xticks
ax.set_yticklabels([])

plt.show()
```
