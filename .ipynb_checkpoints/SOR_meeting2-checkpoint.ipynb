```python
import numpy as np
import glob
import matplotlib.pyplot as plt
import matplotlib
import os
import random
from matplotlib.lines import Line2D

# Matplotlib rc parameters modification

plt.rcParams.update({
  "figure.max_open_warning" : 0,
  "text.usetex"             : True,
  "text.latex.preamble"     : r"\usepackage{amsmath, bm, type1cm}",  # Added \bm for bold math
  "figure.autolayout"       : True,
  "font.family"             : "STIXGeneral",
  "mathtext.fontset"        : "stix",
  "font.size"               : 8,
  "xtick.labelsize"         : 8,
  "ytick.labelsize"         : 8,
  "lines.linewidth"         : 1,
  "lines.markersize"        : 5,
})
#plt.rcParams['text.latex.preamble']= r"\usepackage{amsmath}"
matplotlib.rc('text', usetex=True)
matplotlib.rcParams['text.latex.preamble'] = r'\boldmath'

colors = ['#4a91b5', '#e68139', '#5da258', '#87629b', '#1b9e77']

# reading rig files

# def rigList(rigFile):
#     hashCounter = -4
#     clusterIDs  = []
#     for line in rigFile:
#         if line[0] == '#':
#             hashCounter += 1
#         elif hashCounter >= 0:
#             clusterIDs.append(line.strip())
#     return clusterIDs
def rigList(rigFile):
    hashCounter = -4
    clusterIDs  = []
    temp = []
    for line in rigFile:
        if line[0] == '#':
            hashCounter += 1
            if len(temp) > 0:
                clusterIDs.append(temp)
                temp = []
        elif hashCounter >= 0:
            temp.append(line.strip())
            
    rigClusterIDsList = []
    for ii, sampleList in enumerate(clusterIDs):
        tempList = []
        for kk in range(len(sampleList)):
            tempList.append([int(indx) for indx in sampleList[kk].split(',')])
        rigClusterIDsList.append(tempList)
    return rigClusterIDsList

def particleSizeList(randomSeedFile, sizeRatio, npp = 1000):
    '''
    This function reads the random seed file and creates
    a list of particle sizes. The list index is the particle index.

    Inputs:
    randomSeedFile - location for the random seed file. It contains the particle index and sizes
    sizeRatio      - delta or ar for the case
    npp            - system size
    '''

    if sizeRatio == 1:
        # Painting particles randomly in two colours for monodisperse case.
        particleSize = [1]*(int(npp/2)) + [2]*(int(npp/2))
        random.shuffle(particleSize)
    else:
        particleSize = np.loadtxt(randomSeedFile, usecols = 3) # reading only column 3 which has particle size
        randomSeedFile.close()
    return particleSize


def interactionsList(interactionFile):
    '''
    This function reads the interaction file and creates a nested-list,
    each list inside contains the array of all interaction parameters for
    that timestep.

    Input: interactionFile - the location of the interaction data file
    '''

    hashCounter = 0
    temp        = []
    contactList = [] # list with interaction parameters for each element at each timestep

    fileLines = interactionFile.readlines()[27:] # skipping the comment lines
    for line in fileLines:
        if not line.split()[0] == '#':
            lineList = [float(value) for value in line.split()]
            temp.append(lineList)
        else:
            hashCounter += 1 # checking if counter reaches 7 (7 lines of comments after every timestep data)
            if hashCounter == 7: 
                contactList.append(np.array(temp))
                temp        = []
                hashCounter = 0
    interactionFile.close()
    return contactList

def parametersList(ParametersFile):
    '''
    This function reads the parameters file and creates a nested-list,
    each list inside contains the array of all interaction parameters for
    that timestep.

    Input: ParametersFile - the location of the parameters data file
    '''

    parFile = open(ParametersFile, 'r')

    hashCounter = 0
    temp        = []
    parList     = [] # list with parameters parameters for each element at each timestep

    fileLines = parFile.readlines()[22:] # skipping the comment lines
    for line in fileLines:
        if not line.split()[0] == '#':
            lineList = [float(value) for value in line.split()]
            temp.append(lineList)
        else:
            hashCounter += 1 # checking if counter reaches 7 (7 lines of comments after every timestep data)
            if hashCounter == 7: 
                parList.append(np.array(temp))
                temp        = []
                hashCounter = 0
    parFile.close()
    return parList
```

```python
# Rig large-small analysis (with phim)
# rig particle count

NP    = [1000]
phi   = [0.70,0.71,0.72,0.73,0.74,0.75,0.76] #,0.78,0.79,0.792]
ar    = [1.0, 2.0, 4.0]
phim  = [0.7839319645227314,0.783317545762567,0.7854135245320721,0.7884149253842823,0.8181942616808251]

numRuns = 4
ranSeedFile = "random_seed.dat"

topDir = '/media/rahul/Rahul_2TB/high_bidispersity'
#fig_save_path = "/media/Linux_1TB/City College Dropbox/Rahul Pandare/CUNY/Research/Bidisperse Project/SOR_meeting/figures/"
fig_save_path = "/home/rahul/Dropbox (City College)/CUNY/Research/Bidisperse Project/SOR_meeting/figures/"

cmap = matplotlib.colormaps['viridis_r']
off = 100

line_markers = ['o','^', 's']

for i,npp in enumerate(NP):
    for j, arj in enumerate(ar):
        times = []
        timel = []
        for k, phij in enumerate(phi):
            s = 0
            l = 0
            phir = '{:.3f}'.format(phij) if len(str(phij).split('.')[1]) > 2 else '{:.2f}'.format(phij)
            for ll in range(numRuns):
                workDir = f'{topDir}/NP_{npp}/phi_{phir}/ar_{arj}/Vr_0.5/run_{ll+1}'
                if os.path.exists(workDir):
                    rigFile = open(glob.glob(f'{workDir}/rig_*.dat')[0])
                    rigClusterIDs = rigList(rigFile)
                    numSamples = len(rigClusterIDs[off:])
                    clusterIDs     = [indx for riglist in rigClusterIDs[off:] for indxList in riglist if len(indxList) > 2 for indx in indxList]     
                    ranFile      = open(f'{workDir}/{ranSeedFile}', 'r')
                    particleSize = particleSizeList(ranFile, sizeRatio = ar[j])
    
                    ranFile = open(f'{workDir}/{ranSeedFile}', 'r')
                    lines   = ranFile.readlines()
                    nps     = int(lines[1].split()[1])
                    npl     = int(lines[1].split()[2])
                    
                    s += sum(particleSize[ii] == 1 for ii in clusterIDs)
                    l += sum(particleSize[ii] != 1 for ii in clusterIDs)
            times.append(s/(numSamples*nps*numRuns) if s > 0 else np.nan)
            timel.append(l/(numSamples*npl*numRuns) if l > 0 else np.nan)

        phim2 = [bb/phim[j] for bb in phi]
        plt.plot(phim2, times, linestyle='--', marker=line_markers[j], color=cmap(1/len(ar)), linewidth=1.2)
        plt.plot(phim2, timel, linestyle='--', marker=line_markers[j], color=cmap(2/len(ar)), linewidth=1.2)

color_legend_elements = [
    Line2D([0], [0], color=cmap(2 / len(ar)), lw=2, label=r'\textbf{Large particles}'),
    Line2D([0], [0], color=cmap(1 / len(ar)), lw=2, label=r'\textbf{Small particles}')
    ]

marker_legend_elements = [
        Line2D([0], [0], marker=line_markers[j],lw=0, label=r'$\delta = {:.1f}$'.format(ar[j]), markersize=6)
        for j in range(len(ar))
        ]

legend1 = plt.legend(handles=color_legend_elements, loc='best', title=r"\textbf{Colors}",
                     title_fontsize='10', borderpad=1.05, prop={'size': 10}) #loc=(0.41, 0.64)
legend2 = plt.legend(handles=marker_legend_elements, loc=(0.04, 0.48), title=r"\textbf{Markers}",
                    title_fontsize='10', borderpad=1.05, prop={'size': 12}) #loc=(0.75, 0.62)
plt.gca().add_artist(legend1)

plt.grid(which='both', alpha = 0.2)
plt.xlabel(r'$\phi / \phi_{m}$', fontsize=16, fontweight='bold', style='italic')
plt.ylabel(r'$\textbf{Particle Count}$', fontsize=16, fontweight='bold', style='italic')
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.tight_layout()

figsave = True

if figsave:
    figFormat=".svg"
    plt.savefig(fig_save_path+ "rig_large_small_nps_npl_" +str(npp)+figFormat, bbox_inches="tight",dpi=500)

plt.show()
```

```python
# Rig large-small analysis (with phim) normalised by total NP
## here we take the temporal mean of max of the T (rigidity persistency) 
## rigList needs to be updated to spit out list equal to the length of the run

def rigList(rigFile):
    hashCounter = -4
    clusterIDs  = []
    temp = []
    for line in rigFile:
        if line[0] == '#':
            hashCounter += 1
            if len(temp) > 0:
                clusterIDs.append(temp)
                temp = []
        elif hashCounter >= 0:
            temp.append(line.strip())
            
    rigClusterIDsList = []
    for ii, sampleList in enumerate(clusterIDs):
        tempList = []
        for kk in range(len(sampleList)):
            tempList.append([int(indx) for indx in sampleList[kk].split(',')])
        rigClusterIDsList.append(tempList)
    return rigClusterIDsList
    
def longest_consecutive_ones(arr):
    max_count = 0
    current_count = 0

    for num in arr:
        if num == 1:
            current_count += 1
        else:
            max_count = max(max_count, current_count)
            current_count = 0

    # In case the list ends with a sequence of ones
    return max(max_count, current_count)

NP    = [1000]
phi   = [0.70,0.71,0.72,0.73,0.74,0.75,0.76] #,0.78,0.79,0.792]
ar    = [1.0, 2.0, 4.0]
phim  = [0.7839319645227314,0.783317545762567,0.7854135245320721,0.7884149253842823,0.8181942616808251]

numRuns = 4
ranSeedFile = "random_seed.dat"

topDir = '/media/rahul/Rahul_2TB/high_bidispersity'
fig_save_path = "/home/rahul/Dropbox (City College)/CUNY/Research/Bidisperse Project/SOR_meeting/figures/"

cmap = matplotlib.colormaps['viridis_r']
off = 100

line_markers = ['o','^', 's']

for i,npp in enumerate(NP):
    for j, arj in enumerate(ar):
        times = []
        timel = []
        for k, phij in enumerate(phi):
            maxS = []
            maxL = []
            phir = '{:.3f}'.format(phij) if len(str(phij).split('.')[1]) > 2 else '{:.2f}'.format(phij)
            for ll in range(numRuns):
                workDir = f'{topDir}/NP_{npp}/phi_{phir}/ar_{arj}/Vr_0.5/run_{ll+1}'
                if os.path.exists(workDir):
                    rigFile       = open(glob.glob(f'{workDir}/rig_*.dat')[0])
                    rigClusterIDs = rigList(rigFile)
                    
                    rigMatrix     = np.zeros((npp, len(rigClusterIDs[off:])))
                    ranFile       = open(f'{workDir}/{ranSeedFile}', 'r')
                    particleSize  = particleSizeList(ranFile, sizeRatio = ar[j])
                
                    for ii,riglist in enumerate(rigClusterIDs[off:]):
                        for jj,indxList in enumerate(riglist):
                            for kk,indx in enumerate(indxList):
                                rigMatrix[indx, ii] = 1
                                
                    for il in range(npp):
                        if particleSize[il] ==1:
                            maxS.append(longest_consecutive_ones(rigMatrix[:,il]))
                        else:
                            maxL.append(longest_consecutive_ones(rigMatrix[:,il]))
                            
            times.append(np.mean(maxS) if len(maxS)>0 else np.nan) 
            timel.append(np.mean(maxL) if len(maxL)>0 else np.nan)

        phim2 = [bb/phim[j] for bb in phi]
        plt.plot(phim2, times, linestyle='--', marker=line_markers[j], color=cmap(1/len(ar)), linewidth=1.2)
        plt.plot(phim2, timel, linestyle='--', marker=line_markers[j], color=cmap(2/len(ar)), linewidth=1.2)

color_legend_elements = [
    Line2D([0], [0], color=cmap(1 / len(ar)), lw=2, label=r'\textbf{Small particles}'),
    Line2D([0], [0], color=cmap(2 / len(ar)), lw=2, label=r'\textbf{Large particles}')
    ]

marker_legend_elements = [
        Line2D([0], [0], marker=line_markers[j],lw=0, label=r'$\delta = {:.1f}$'.format(ar[j]), markersize=6)
        for j in range(len(ar))
        ]

legend1 = plt.legend(handles=color_legend_elements, loc=(0.11, 0.56), 
                     title=r"\textbf{Colors}", borderpad=1.05, prop={'size': 10})
legend2 = plt.legend(handles=marker_legend_elements, loc=(0.55, 0.54), 
                     title=r"\textbf{Markers}", borderpad=1.05, prop={'size': 10})

plt.gca().add_artist(legend1)
plt.grid(which='both', alpha=0.2)
plt.xlabel(r'$\phi / \phi_{m}$', fontsize=16, fontweight='bold', style='italic')
plt.ylabel('max(t)', fontsize=16, fontweight='bold')
plt.tight_layout()

figsave = True

if figsave:
    figFormat=".svg"
    plt.savefig(fig_save_path+ "frig_large_small_xx" +str(npp)+figFormat, bbox_inches="tight",dpi=500)

plt.show()
```

```python
# Rig large-small analysis (with phim) normalised by total NP

def rigList(rigFile):
    hashCounter = -4
    clusterIDs  = []
    temp = []
    for line in rigFile:
        if line[0] == '#':
            hashCounter += 1
            if len(temp) > 0:
                clusterIDs.append(temp)
                temp = []
        elif hashCounter >= 0:
            temp.append(line.strip())
            
    rigClusterIDsList = []
    for ii, sampleList in enumerate(clusterIDs):
        tempList = []
        for kk in range(len(sampleList)):
            tempList.append([int(indx) for indx in sampleList[kk].split(',')])
        rigClusterIDsList.append(tempList)
    return rigClusterIDsList
    
def lengths_of_consecutive_ones_greater_than_two(arr):
    lengths = []
    current_count = 0

    for num in arr:
        if num == 1:
            current_count += 1
        else:
            if current_count > 2:
                lengths.append(current_count)
            current_count = 0

    # If the list ends with a sequence of ones
    if current_count > 2:
        lengths.append(current_count)

    return lengths

#########

topDir = '/media/rahul/Rahul_2TB/high_bidispersity'
fig_save_path = "/home/rahul/Dropbox (City College)/CUNY/Research/Bidisperse Project/SOR_meeting/figures/"


numRuns = 4
ranSeedFile = "random_seed.dat"
cmap = matplotlib.colormaps['viridis_r']
off = 100

line_markers = ['o','^', 's']

def rigPers(phi, ar, fsize = (4,6), hidey = False, figsave = False):
    maxS = []
    maxL = []
    if ar == 4:
        phi = 0.78
        
    for ll in range(numRuns):
        workDir = f'{topDir}/NP_{1000}/phi_{phi}/ar_{ar}/Vr_0.5/run_{ll+1}'
        if os.path.exists(workDir):
            rigFile       = open(glob.glob(f'{workDir}/rig_*.dat')[0])
            rigClusterIDs = rigList(rigFile)
            
            rigMatrix     = np.zeros((npp, len(rigClusterIDs[off:])))
            ranFile       = open(f'{workDir}/{ranSeedFile}', 'r')
            particleSize  = particleSizeList(ranFile, sizeRatio = ar)
        
            for ii,riglist in enumerate(rigClusterIDs[off:]):
                for jj,indxList in enumerate(riglist):
                    for kk,indx in enumerate(indxList):
                        rigMatrix[indx, ii] = 1
                        
            for il in range(npp):
                if particleSize[il] ==1:
                    maxS.append(lengths_of_consecutive_ones_greater_than_two(rigMatrix[:,il]))
                else:
                    maxL.append(lengths_of_consecutive_ones_greater_than_two(rigMatrix[:,il]))
                
    flat_s = [gg for sublist in maxS for gg in sublist]
    flat_l = [gg for sublist in maxL for gg in sublist]
    
    all_values = np.unique(np.concatenate([flat_s, flat_l]))
    
    counts1 = np.array([flat_s.count(val) for val in all_values])
    counts2 = np.array([flat_l.count(val) for val in all_values])
    counts1_normalized = counts1 / len(flat_s)
    counts2_normalized = counts2 / len(flat_l)
    
    bar_width = 0.4
    bar_positions1 = np.arange(len(all_values))
    bar_positions2 = bar_positions1 + bar_width  # Shift positions for the second set
    
    plt.figure(figsize=fsize)
    plt.bar(bar_positions1, counts1_normalized, width=bar_width, color='blue', label=fr'\textbf{{Small particles}}')
    plt.bar(bar_positions2, counts2_normalized, width=bar_width, color='red', label=fr'\textbf{{Large particles}}')
    
    latex_labels = [fr'$\textbf{{{val}}}$' for val in all_values]
    plt.xticks(bar_positions1 + bar_width / 2, latex_labels, fontsize=18)
    plt.yticks(fontsize=16)
   
    plt.xlabel(r'$\textbf{T}$', fontsize=18, fontweight='bold', style='italic')
    plt.ylabel(r'$\textbf{Relative Frequency}$', fontsize=18, fontweight='bold', style='italic')
    plt.title(fr'$\delta = {ar}$', fontsize=18, pad =10)
    plt.legend(loc='upper right', fontsize=16, labelspacing=1.5, borderpad=1.2)
    plt.grid(which='both', alpha=0.2)
    plt.yscale('log')

    if hidey:
        plt.gca().set_yticklabels([])
        plt.ylabel('')
        plt.legend().set_visible(False)
    
    plt.tight_layout()
    
    if figsave:
        figFormat=".svg"
        plt.savefig(fig_save_path+ "rigPers_phi_" + str(phi)+ "_ar_" + str(ar) + "_np_" +str(1000)+figFormat, bbox_inches="tight", dpi=500)
    
    plt.show()
```

```python
# maron pierce plot

from prettytable import PrettyTable
import warnings
warnings.filterwarnings("ignore", message="Mean of empty slice.", category=RuntimeWarning)
warnings.filterwarnings("ignore", message="invalid value encountered in scalar divide", category=RuntimeWarning)

from scipy.optimize import curve_fit

def maronpierce(x,c, k):
    return c*(1 - (x/k)) ** -2

phi = [0.70,0.71,0.72,0.73,0.74,0.75,0.76,0.77,0.78,0.79, 0.792]
ar  = [1.0, 1.4, 1.8, 2.0, 4.0]

phi_vals  = phi
arr       = ar
numRuns   = 4

rerr    = [] # r2 error
phim    = [] # max phi calculated from fitting
c_list  = [] # free parameter c in the model
k_guess = [.785,.785,.785,.788,.818] # initial guess for max phi (phim)

topDir        = '/media/rahul/Rahul_2TB/high_bidispersity'
fig_save_path = "/home/rahul/Dropbox (City College)/CUNY/Research/Bidisperse Project/SOR_meeting/figures/"

#topDir        = "/Volumes/Rahul_2TB/high_bidispersity/"
#fig_save_path = "/Users/rahul/City College Dropbox/Rahul Pandare/CUNY/Research/Bidisperse Project/SOR_meeting/figures/"

npp = 1000
y   = 3    #rheological parameter  -viscosity
off = 100

visclist = []
philist = []
cmap = matplotlib.colormaps['viridis_r'] #color scheme
line_markers = ['o', 's', '^', '*', 'p']
colors = ['#4a91b5', '#e68139', '#5da258', '#87629b', '#1b9e77']
plt.clf()
plt.figure()
for j, arj in enumerate(arr):
    visc = []
    for i, phii in enumerate(phi_vals):
        dat = []
        phir = '{:.3f}'.format(phii) if len(str(phii).split('.')[1])>2 else '{:.2f}'.format(phii)
        for l in range(numRuns):
            workDir = f'{topDir}/NP_1000/phi_{phir}/ar_{ar[j]}/Vr_0.5/run_{l+1}'
            if not os.path.exists(workDir):
                continue
            file_name = glob.glob(f'{workDir}/data_*.dat')[0]
            with open(f'{file_name}','r') as file:
                data_array = np.loadtxt(file) 
                data = data_array[:, 3][off:]
                dat.append(np.mean(data))
        visc.append(np.mean(dat))
       
    # points to fit
    viscf = [x for x in visc if not np.isnan(x)] # filtered dat list to remove 'nan'
    phif  = [phi_vals[i] for i in range(len(viscf)) if not np.isnan(viscf[i])] # filtered phi list to remove phi values for cases that dont exist
    visclist.append(viscf)
    philist.append(phif)
    
    # curve fit
    popt, pcov   = curve_fit(maronpierce, phif, viscf,[1, k_guess[j]])
    c_opt, k_opt = popt
    phim.append(k_opt)
    plt.plot(phi_vals, visc,linestyle='None',marker=line_markers[j], label=r'$\delta = {:.1f}$ ($\phi_m = {:.3f}$)'.format(arj,k_opt), 
             color=colors[j],linewidth=1.2, markersize=6.5)
    #print(f"Optimal value for c (del = {arj:.1f}): {c_opt:.4f}")
    c_list.append(c_opt)

    # plotting for model
    xx = np.linspace(0.70, .796, 100)
    yy = maronpierce(xx,c_opt,k_opt)
    plt.plot(xx, yy,color = colors[j], linewidth=1.5) #cmap((j+1)/len(arr)))

    #coefficient of determinantion (r^2)
    r2 = True
    if r2:
        # calculating the r2 for each curve
        y_predicted = [maronpierce(i, *popt) for i in phif]
        residuals = [a - b for a, b in zip(viscf, y_predicted)]
        TSS = np.sum((viscf - np.mean(viscf))**2)
        RSS = np.sum([i**2 for i in residuals])
        R_squared = 1 - (RSS / TSS)
        rerr.append(R_squared)
        #print("R-squared value:", R_squared)

# printing table
table = PrettyTable()
column_names = ['delta', 'phi_m', "C", "R2"]
table.add_column(column_names[0], arr)
table.add_column(column_names[1], ["{:.3f}".format(x) for x in phim])
table.add_column(column_names[2], ["{:.3f}".format(x) for x in c_list])
table.add_column(column_names[3], ["{:.3f}".format(x) for x in rerr])
print(table)

print("\n R-squared mean:", ("{:.3f}".format(np.mean(rerr)))) #R-squared value

# plot parameters
plt.grid(which='Both', alpha=0.2)
plt.xlim(.698, .794)
#plt.yscale('log')
#plt.ylim(80, 2.3e3) 
plt.ylim(30,2150)
plt.xlabel(r'$\phi$', fontsize=18,fontstyle='italic')
plt.ylabel(r'$\mathit{\eta_{r}}$', fontsize=16,fontstyle='italic', labelpad=10)

plt.xticks(fontsize=13)
plt.yticks(fontsize=12)
plt.legend(loc='upper left', fontsize=11, labelspacing=1.5,borderpad=1.2)
plt.tight_layout()

figsave=True

if figsave:
    figFormat=".svg"
    plt.savefig(fig_save_path+ "Maron-Pierce" +str(npp)+figFormat, bbox_inches="tight",dpi=500)

plt.show()
```

```python
## contacts

import os
import matplotlib                   # type: ignore
import numpy             as     np  # type: ignore
import matplotlib.pyplot as     plt # type: ignore
import random

def updateContAngle(theta, contList):
    '''
    This function to calulated the complementary contact angle and append to 
    relevant list (contList)

    Inputs:
    theta    - primary contact angle. Calculated by arctan2(nz,nx)
    contList - list where the angle appends. This list contains all the contact angle for one case, all runs
    '''
    contList.append(theta)
    if theta < 0:
        contList.append(np.pi - abs(theta))
    else:
        contList.append(theta - np.pi)

def particleSizeList(randomSeedFile, sizeRatio, npp = 1000):
    '''
    This function reads the random seed file and creates
    a list of particle sizes. The list index is the particle index.

    Inputs:
    randomSeedFile - location for the random seed file. It contains the particle index and sizes
    sizeRatio      - delta or ar for the case
    npp            - system size
    '''

    if sizeRatio == 1:
        # Painting particles randomly in two colours for monodisperse case.
        particleSize = [1]*(int(npp/2)) + [2]*(int(npp/2))
        random.shuffle(particleSize)
    else:
        particleSize = np.loadtxt(randomSeedFile, usecols = 3) # reading only column 3 which has particle size
        randomSeedFile.close()
    return particleSize

def interactionsList(interactionFile):
    '''
    This function reads the interaction file and creates a nested-list,
    each list inside contains the array of all interaction parameters for
    that timestep.

    Input: interactionFile - the location of the interaction data file
    '''

    hashCounter = 0
    temp        = []
    contactList = [] # list with interaction parameters for each element at each timestep

    fileLines = interactionFile.readlines()[27:] # skipping the comment lines
    for line in fileLines:
        if not line.split()[0] == '#':
            lineList = [float(value) for value in line.split()]
            temp.append(lineList)
        else:
            hashCounter += 1 # checking if counter reaches 7 (7 lines of comments after every timestep data)
            if hashCounter == 7: 
                contactList.append(np.array(temp))
                temp        = []
                hashCounter = 0
    interactionFile.close()
    return contactList 
    
def angularContactsAllSizePair(phii, arr, figsave = False):
    '''
    This function plots histogram of contact angles for one phi value and ar value for all size pairs.
    The plots would be Z/Z0 vs theta, it will have 4 curves ( 3 for different size pairs and 1 total).

    Inputs:
    phii - phi value
    arr  - ar value
    '''

    plt.clf()
    #cmap = matplotlib.colormaps['viridis_r'] #color scheme
    colors = ['#4a91b5', '#e68139', '#5da258', '#D80000']
    linestyles = ['None', 'None', 'None', '--']
    
    #colors = ['#4a91b5', '#e68139', '#5da258', '#87629b', '#1b9e77']
    #linestyles = ['None', 'None', 'None', 'None']
    
    npp = 1000 # no. of particles
    off = 100  # steady state cut off

    phii = '{:.3f}'.format(phii) if len(str(phii).split('.')[1])>2 else '{:.2f}'.format(phii)

    contactAngleAllRuns   = [[] for _ in range(run[npp])] # empty nested list for all four runs 
    sizePair              = ['Total', 'SS', 'SL', 'LL']
    line_markers = ['o', 's', '^', 'p']
    
    dataname = topDir+'NP_'+str(npp)+'/phi_'+phii+'/ar_'+str(arr)+'/Vr_0.5'
    if os.path.exists(dataname):
        for l in range (run[npp]):
            ranFile      = open(f'{dataname}/run_{l+1}/{ranSeedFile}', 'r')
            particleSize = particleSizeList(ranFile, sizeRatio = arr) 

            interFile   = open(f'{dataname}/run_{l+1}/{intFile}', 'r')
            contactList = interactionsList(interFile) 

            contactAngleTotal = [[] for _ in range(run[npp])]
            for sampleList in contactList[off:]:
                timeStepContactAngle = [[] for _ in range(run[npp])] # all contact angles at each timestep
                for i in range (sampleList.shape[0]):
                    particleSize1 = particleSize[int(sampleList[i, 0])]
                    particleSize2 = particleSize[int(sampleList[i, 1])]
                    contState     = int(sampleList[i, 10]) # Contact state
                    thetaRad      = np.arctan2(sampleList[i,4],sampleList[i,2])

                    if contState == 2:
                        thetaRad = np.arctan2(sampleList[i,4], sampleList[i,2]) # contact angle = tan^-1(nz/nx) ; nz, nx: unit vector in z and x
                        updateContAngle(thetaRad, timeStepContactAngle[0])

                        # Small-Small
                        if (particleSize1 == particleSize2 == 1):
                            updateContAngle(thetaRad, timeStepContactAngle[1])

                        # Small-Large
                        elif (particleSize1 != particleSize2):
                            updateContAngle(thetaRad, timeStepContactAngle[2])

                        # Large-Large
                        elif (particleSize1 == particleSize2 > 1):
                            updateContAngle(thetaRad, timeStepContactAngle[3])

                for m in range(len(contactAngleTotal)):
                    contactAngleTotal[m].append(timeStepContactAngle[m]) # nested list of all contact angles

            for n in range(len(contactAngleAllRuns)):
                contactAngleList = [i for sub in contactAngleTotal[n] for i in sub]
                contactAngleAllRuns[n].append(contactAngleList)

    # Loop for plotting histogram for each size pair 
    for ii in range(len(contactAngleAllRuns)):
    #for ii in range(1):
        allContactAngles  = [i for sub in contactAngleAllRuns[ii] for i in sub]
        counts, bin_edges = np.histogram(allContactAngles, bins = 72, density = True)
        bin_centers       = (bin_edges[:-1] + bin_edges[1:]) / 2
        countsAvg         = np.mean(counts)
        plt.plot(bin_centers, counts/countsAvg, linestyle = linestyles[ii], 
                 marker=line_markers[ii], markersize = 5.5, label=fr'\textbf{{{sizePair[ii]}}}', color = colors[ii])

    xticks       = [-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi]
    xtick_labels = [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$']
    #plt.xticks(xticks, xtick_labels, fontsize = 14)
    plt.title(fr'$\mathbf{{\delta = {arr}}}$', fontsize = 24)
    #plt.xlabel(r'$\theta$',  fontsize = 16)
    #plt.ylabel(r'$Z/\bar{Z}$', fontsize = 16)
    
    #plt.grid(True, alpha=0.4)
    #plt.gca().xaxis.set_ticklabels([])


    x_grid_points = [-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi]  
    plt.gca().set_xticks(x_grid_points)
    plt.gca().xaxis.set_ticklabels([])

    # Display grid only at the specified x-tick locations
    plt.grid(True, alpha=0.4, axis='both', which='both')
    
    #plt.xticks(fontsize = 14)
    
    #plt.yticks(fontsize = 18)
    #plt.tick_params(axis='y', which='both', left=True, right=False, labelleft=False)  # Disable y ticks but keep grid
    
    
    if arr != 1.4:
        plt.tick_params(axis='y', which='both', left=True, right=False, labelleft=False)  # comment this for 1.0 
    else:
        plt.ylabel(r'$g(\theta)$', fontsize = 16, labelpad=5) # comment this for 2.0 and 4.0
        plt.yticks(fontsize = 18)
    
    plt.ylim(0, 2.5)
    
    #if arr == 2:
        #plt.legend(fontsize = 16, labelspacing = 1, borderpad = 1, markerscale=1.2)

    plt.tight_layout()
    if figsave:
        figFormat     = ".svg"
        plt.savefig(fig_save_path + '/angularContactsAllSizePair_red_NP_' + str(1000) + '_phi_' + str(phii) + '_ar_' + str(arr) + figFormat, bbox_inches = "tight", dpi = 500)

    plt.show()
```

```python
# Below are the functions used in the above histogram functions. Esessentially to shorten it.
import os
import matplotlib                   # type: ignore
import numpy             as     np  # type: ignore
import matplotlib.pyplot as     plt # type: ignore
import random
import matplotlib.legend as mlegend

def updateContStress(theta, binCentersList, stressList, stressVal):
    '''
    This function is used to calulated the complementary contact angle and 
    sum-append the stress for that contact to relevant list

    Inputs:
    theta          - primary contact angle. Calculated by arctan2(nz,nx)
    binCentersList - centers of bins
    forceList      - list of total Stress for each bin. The stress appends here at the relevant bin
    forceVal       - the value of norm of the normal contact stress to be sum-appended to 'forceList'
    '''
    
    bin_Center = int(np.floor(len(binCentersList)/2))
    bin_Index = np.argmin(np.abs(binCentersList - theta))
    stressList[bin_Index] += stressVal

    if bin_Index != bin_Center:
        bin_Index2 = (bin_Center + 1) + bin_Index if bin_Index < bin_Center else bin_Index - (bin_Center + 1)
        stressList[bin_Index2] += stressVal

def particleSizeList(randomSeedFile, sizeRatio, npp = 1000):
    '''
    This function reads the random seed file and creates
    a list of particle sizes. The list index is the particle index.

    Inputs:
    randomSeedFile - location for the random seed file. It contains the particle index and sizes
    sizeRatio      - delta or ar for the case
    npp            - system size
    '''

    if sizeRatio == 1:
        # Painting particles randomly in two colours for monodisperse case.
        particleSize = [1]*(int(npp/2)) + [2]*(int(npp/2))
        random.shuffle(particleSize)
    else:
        particleSize = np.loadtxt(randomSeedFile, usecols = 3) # reading only column 3 which has particle size
        randomSeedFile.close()
    return particleSize

def interactionsList(interactionFile):
    '''
    This function reads the interaction file and creates a nested-list,
    each list inside contains the array of all interaction parameters for
    that timestep.

    Input: interactionFile - the location of the interaction data file
    '''

    hashCounter = 0
    temp        = []
    contactList = [] # list with interaction parameters for each element at each timestep

    fileLines = interactionFile.readlines()[27:] # skipping the comment lines
    for line in fileLines:
        if not line.split()[0] == '#':
            lineList = [float(value) for value in line.split()]
            temp.append(lineList)
        else:
            hashCounter += 1 # checking if counter reaches 7 (7 lines of comments after every timestep data)
            if hashCounter == 7: 
                contactList.append(np.array(temp))
                temp        = []
                hashCounter = 0
    interactionFile.close()
    return contactList

def parametersList(ParametersFile):
    '''
    This function reads the parameters file and creates a nested-list,
    each list inside contains the array of all interaction parameters for
    that timestep.

    Input: ParametersFile - the location of the parameters data file
    '''

    parFile = open(ParametersFile, 'r')

    hashCounter = 0
    temp        = []
    parList     = [] # list with parameters parameters for each element at each timestep

    fileLines = parFile.readlines()[22:] # skipping the comment lines
    for line in fileLines:
        if not line.split()[0] == '#':
            lineList = [float(value) for value in line.split()]
            temp.append(lineList)
        else:
            hashCounter += 1 # checking if counter reaches 7 (7 lines of comments after every timestep data)
            if hashCounter == 7: 
                parList.append(np.array(temp))
                temp        = []
                hashCounter = 0
    parFile.close()
    return parList

def angularStressAllSizePair(phii, arr, figsave = False):
    '''
    This function plots angular Stress for one phi value and ar value for all size pairs.
    The plots would be sigma_c/sigma_0 vs theta, it will have 4 curves ( 3 for different size pairs and 1 total).

    Inputs:
    phii - phi value
    arr  - ar value
    '''

    plt.clf()
    cmap = matplotlib.colormaps['viridis_r'] #color scheme

    npp = 1000 # no. of particles
    off = 100  # steady state cut off

    angleBins  = np.linspace(-np.pi, np.pi,72)
    binCenters = (angleBins[:-1] + angleBins[1:])/2

    phii = '{:.3f}'.format(phii) if len(str(phii).split('.')[1]) > 2 else '{:.2f}'.format(phii)

    stressBin_Sum = [[0]*len(binCenters) for _ in range(run[npp])] # empty nested list for all four runs 
    contCount    = [0]*run[npp]
    sizePair     = ['All', 'SS', 'SL', 'LL']
    line_markers = ['o', 's', '^', 'p']
    
    dataname     = topDir + 'NP_' + str(npp) + '/phi_' + phii + '/ar_' + str(arr) + '/Vr_0.5'
    if os.path.exists(dataname):
        for l in range (run[npp]):
            # List of particle size. List index is particle index.
            ranFile      = open(f'{dataname}/run_{l+1}/{ranSeedFile}', 'r')
            particleSize = particleSizeList(ranFile, sizeRatio = arr) 

            # Nested list of all interaction params, each list is one timestep.
            interFile   = open(f'{dataname}/run_{l+1}/{intFile}', 'r')
            contactList = interactionsList(interFile) 
           
            for sampleList in contactList[off:]:
                for i in range (sampleList.shape[0]):
                    particleSize1 = particleSize[int(sampleList[i,0])]
                    particleSize2 = particleSize[int(sampleList[i,1])]
                    contState     = int(sampleList[i,10]) # Contact state
                    contForce     = sampleList[i,11] # Norm of normal contact force
                    thetaRad      = np.arctan2(sampleList[i,4], sampleList[i,2])

                    # contact length calc
                    radSum           = particleSize1 +particleSize2
                    dimensionlessGap = sampleList[i, 5]
                    particleOverlap  = -(dimensionlessGap+1)*radSum/2
                    contStress       = contForce/particleOverlap
                    
                    if contState == 2:
                        # For all particles.
                        updateContStress(thetaRad, binCenters, stressBin_Sum[0], contStress)
                        contCount[0] += 2
                        
                        # Small-Small.
                        if (particleSize1 == particleSize2 == 1):
                            updateContStress(thetaRad, binCenters, stressBin_Sum[1], contStress)
                            contCount[1] += 2

                        # Small-Large
                        elif (particleSize1 != particleSize2):
                            updateContStress(thetaRad, binCenters, stressBin_Sum[2], contStress)
                            contCount[2] += 2

                        # Large-Large
                        elif (particleSize1 == particleSize2 > 1):
                            updateContStress(thetaRad, binCenters, stressBin_Sum[3], contStress)
                            contCount[3] += 2

    # Loop for plotting each size pair 
    for ii in range(len(stressBin_Sum)):
        stressAvg     = np.mean(stressBin_Sum[ii])
        stressDensity = [i/stressAvg for i in stressBin_Sum[ii]]
        #forceDensity = [i/contCount[ii]/binWidth for i in forceBin_Sum[ii]]
        plt.plot(binCenters, stressDensity, line_markers[ii], markersize = 4, label=fr'\textbf{{{sizePair[ii]}}}', color = cmap((ii+1)/len(stressBin_Sum))) 

    xticks       = [-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi]
    xtick_labels = [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$']
    plt.xticks(xticks, xtick_labels)
    #plt.title(fr'$\mathbf{{\delta = {arr}}}$', fontsize = 18)
    plt.xlabel(r'$\theta$',  fontsize = 16)
    #plt.ylabel(r'$\sigma_c/\sigma_{0}$', fontsize = 16)

    plt.xticks(fontsize = 14)
    plt.yticks(fontsize = 14)
    #plt.legend(fontsize = 13, labelspacing = 1.2, borderpad = 1.2, markerscale=1.5)

    plt.grid(which = 'Both', alpha = 0.2)
    plt.tight_layout()
    
    if figsave:
        figFormat     = ".svg"
        plt.savefig(fig_save_path + '/angularStressAllSizePair_NP_' + str(npp) + '_phi_' + str(phii) + '_ar_' + str(arr) + figFormat, bbox_inches = "tight", dpi = 500)

    plt.show()
```

```python
## Znet

# plotting different contacts in the system for a phi value
# plotting small-small, large-large and small-large contacts on one plot

#topDir   = "/Volumes/Rahul_2TB/high_bidispersity/"
topDir      = "/media/rahul/Rahul_2TB/high_bidispersity/"
fig_save_path = "/media/Linux_1TB/City College Dropbox/Rahul Pandare/CUNY/Research/Bidisperse Project/SOR_meeting/figures/"

contFile = 'contacts.txt'

NP       = [1000]

run      = {500:8,1000:4,2000:2,4000:1}

phi      = [0.76]

ar       = [1.0, 1.4, 1.8, 2.0, 4.0]

off = 100

contType   = ['All contacts', 'S-S contacts', 'S-L contacts', 'L-L contacts']
line_markers = ['o', 's', '^', 'p']
colors = ['#4a91b5', '#e68139', '#5da258', '#87629b', '#1b9e77']
#linestyles = ['None', '--', 'None', 'None']

#cmap = matplotlib.colormaps['viridis_r'] #color scheme

for ii in range(len(NP)):
    for j in range(len(phi)):
        phir       = '{:.3f}'.format(phi[j]) if len(str(phi[j]).split('.')[1])>2 else '{:.2f}'.format(phi[j])
        
        conList    = [[] for _ in range(len(contType))]
        conStdList = [[] for _ in range(len(contType))]
        plt.clf()
        for k in range(len(ar)):
            if ar[k] == 4.0:
                phir = '0.78' 
            dataname=topDir+'NP_'+str(NP[ii])+'/phi_'+phir+'/ar_'+str(ar[k])+'/Vr_0.5'
            if os.path.exists(dataname):
                temp    = [[] for _ in range(len(contType))]
                tempStd = [[] for _ in range(len(contType))]
                
                for l in range (run[NP[ii]]):
                    conFile  = open(f'{dataname}/run_{l+1}/{contFile}', 'r')
                    contData = np.loadtxt(conFile) # reading only column 3 which has particle size
                    conFile.close()
                    
                    for ik in range(len(temp)):
                        ZnetCont = [ a / b for a, b in zip(contData[off:,ik*2], contData[off:,(ik*2)+1]) if b != 0 ]
                        temp[ik].append(np.mean(ZnetCont))
                        tempStd[ik].append(np.var(ZnetCont))
                    
                for ij in range(len(temp)):
                    conList[ij].append(np.mean(temp[ij]))
                    conStdList[ij].append(np.sqrt(np.mean(tempStd[ij])))
        
        for il in range(len(conList)):
            plt.errorbar(
                ar, conList[il], yerr=conStdList[il], 
                linestyle='--', marker=line_markers[il], markersize=6.5, 
                label=fr'\textbf{{{contType[il]}}}', 
                color=colors[il],  # fixed length calculation for cmap
                capsize=2, linewidth=1.2, elinewidth=0.5
            )
        
        plt.grid(which='Both', alpha=0.2)

        plt.title(fr"$\phi/\phi_{{m}} \approx {0.95}$", fontsize=19, pad=10)
        plt.xlabel(r'$\delta$', fontsize=16,fontstyle='italic')
        plt.ylabel(r'$\langle Z_{net} \rangle$', fontsize=14,fontstyle='italic')

        plt.xticks(fontsize=14)
        plt.yticks(fontsize=14)
        #plt.legend(fontsize=10, ncols =2, labelspacing=1.2,borderpad=1.2, loc=(0.05, 0.60)) #loc=(0.65, 0.66)

        #plt.tight_layout()
        if True:
            figFormat=".svg"
            #fig_save_path = "/Users/rahul/City College Dropbox/Rahul Pandare/CUNY/Research/Bidisperse Project/analysis/figures/contacts"
            plt.savefig(fig_save_path+'Znet_NP_'+str(1000)+'_phim_0.95_'+figFormat, bbox_inches="tight",dpi=500)

        plt.show()
        #plt.clf()
```
